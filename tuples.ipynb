```python
#Convert a string into a tuple of characters.
my_string="sowjanya"
my_tuple=tuple(my_string)
print(my_tuple)
```

```python
#Merge two tuples into a single tuple.
tuple1=(1,2,3,4,5,6)
tuple2=('a','b','c','d')
tuple3=tuple1+tuple2
print(tuple3)
```

```python
#Iterate over a tuple and perform an operation on each element.
my_tuple=(1,2,3,4,5)
for element in my_tuple:
  result=element*2
  print(result)
```

```python
#Convert a tuple into a list
my_tuple=(1,2,3,4,5,6)
my_list=list(my_tuple)
print(my_list)
```

```python
#Reverse the order of elements in a tuple.
my_tuple=(1,2,3,4,5)
reversed_tuple=my_tuple[::-1]
print(reversed_tuple)
```

```python
#Find the maximum and minimum values in a tuple of numbers.
my_tuple=(1,2,3,4,5,6,7,8,9)
maximum_value=max(my_tuple)
minimum_value=min(my_tuple)
print(maximum_value)
print(minimum_value)
```

```python
#Check if all elements in a tuple satisfy a specific condition
my_tuple=(2,4,6,8,10, 11)
all_even=any(element%2==0 for element in my_tuple)
print(all_even)
```

```python
#Slice a tuple to extract a subset of elements.
my_tuple=(1,2,3,4,5,6,7)
subset_tuple=my_tuple[2:6]
print(subset_tuple)
```

```python
#Create a tuple with repeated elements and remove duplicates.
my_tuple=(1,1,2,3,3,4,4,5,6,5,7)
unique_tuple=tuple(set(my_tuple))
print(unique_tuple)
print(set(my_tuple))
```

```python
#Sort a tuple of strings in alphabetical order.
my_tuple=("my","are","yes","no")
sorted_tuple=tuple(sorted(my_tuple))
print(sorted_tuple)
print(sorted(my_tuple))
```

```python
#Create a tuple of numbers and find the sum of all the elements.
my_tuples=(1,2,3,4,5,6,7)
sum_of_elements=sum(my_tuples)
print(sum_of_elements)
```

```python
#Count the number of occurrences of a specific element in a tuple.
my_tuple=(1,2,2,2,3,4,5,2,2,6,2,7,2)
element=2
count=my_tuple.count(element)
print(count)
```

```python
#Unpack a tuple into individual variables.
my_tuple=(1,2,3,4)
a,b,c,d=my_tuple
print(a)
print(b)
print(c)
print(d)
tuple1=(1,2,3,4,5,6)
tuple2=(a,b,c,d)
tuple3=tuple1+tuple2
print(tuple3)
```

```python
#Create a nested tuple containing multiple tuple
nested_tuple=((1,2,3),('a','b','c'),('apple','mango'))
print(nested_tuple)
```
