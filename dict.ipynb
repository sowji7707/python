```python
#Sort the keys in a dictionary in ascending order.
my_dict={"name":"sowjanaya","age":"32","city":"vijayawada"}
sorted_keys = sorted(my_dict.keys())

for key in sorted_keys:
    print(key, my_dict[key])
```

```python
#Create a dictionary from two lists, one containing keys and the other containing values.
keys = ['key1', 'key2', 'key3']
values = ['value1', 'value2', 'value3']

my_dict = dict(zip(keys, values))
print(my_dict)
```

```python
#Iterate over both the keys and values in a dictionary.

my_dict={"name":"sowjanaya","age":"32","city":"vijayawada"}
for key, value in my_dict.items():
    print(key, value)
```

```python
#Iterate over the values in a dictionary.
my_dict={"name":"sowjanaya","age":"32","city":"vijayawada"}
for value in my_dict.values():
  print(value)
```

```python
#Iterate over the keys in a dictionary.
my_dict={"name":"sowjanaya","age":"32","city":"vijayawada"}

for key in my_dict.keys():
    print(key)
```

```python
#Create a dictionary with predefined key-value pairs.
my_dict={"name":"sowjanaya","age":"32","city":"vijayawada"}
print(my_dict)
```

```python
#Get the length of a dictionary.
my_dict={"name":"sowjanaya","age":"32","city":"vijayawada"}
length=len(my_dict)
print(length)
```

```python
#Clear all the key-value pairs from a dictionary.
my_dict={"name":"sowjanaya","age":"32","city":"vijayawada"}
my_dict.clear()
print(my_dict)
```

```python
#Concatenate two dictionaries.
my_dict1={"name":"sowjanaya","age":"32","city":"vijayawada"}
my_dict2={"father":"sambi reddy"}
my_dict1.update(my_dict2)
print(my_dict1)
```

```python
#Update the value associated with a specific key in a dictionary
my_dict={"name":"sowjanaya","age":"32","city":"vijayawada"}
my_dict["age"]="34"
print(my_dict)
```

```python
#Remove a key-value pair from a dictionary.
my_dict={"name":"sowjanaya","age":"32","city":"vijayawada"}
del my_dict["age"]
print(my_dict)
```

```python
#Retrieve all the values from a dictionary.
my_dict={"name":"sowjanaya","age":"32","city":"vijayawada"}
values=my_dict.values()
print(values)
```
